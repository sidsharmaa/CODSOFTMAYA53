```python
#create a function that seperates the class of transaction between fraud and legit transactions

def split_data_by_class(creditCard):
    
    legit = creditCard[creditCard["Class"] == 0]
    fraud = creditCard[creditCard["Class"] == 1]
    return legit, fraud

# Example usage:
legit_df, fraud_df = split_data_by_class(creditCard)

```

```python
# importing the model building libraries 

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression 
from sklearn.metrics import accuracy_score


# Importing the train_test_split function from sklearn.model_selection module to split data into training and testing sets.
# from sklearn.model_selection import train_test_split

# Importing the LogisticRegression class from sklearn.linear_model module to create a logistic regression model.
# from sklearn.linear_model import LogisticRegression

# Importing the accuracy_score function from sklearn.metrics module to evaluate the accuracy of the model predictions.
# from sklearn.metrics import accuracy_score
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25, random_state = 5)


# Using the train_test_split function to split the feature matrix (x) and target variable (y) into training and testing sets.
# The test_size parameter specifies the proportion of the dataset to include in the test split (here, 25%).
# The random_state parameter sets the seed for random number generation to ensure reproducibility of results.
# x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=5)
```

```python
# Creating a Logistic Regression model object

# Instantiating a LogisticRegression object from the LogisticRegression class in scikit-learn.
model = LogisticRegression()
```
